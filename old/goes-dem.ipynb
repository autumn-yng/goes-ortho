## GOES-16 & DEM

The GOES-R satellites, like other geostationary Earth observing satellites, provide regular observations of nearly an entire hemisphere of Earth. However, because they capture images of such a large area from a single perspective, the local view angle (the angle above the horizon at each point on the ground that the satellite would appear in the sky) varies across the entire scene. Terrain variations across Earth's surface are therefore subject to parallax, especially at the greatest of these view angles.

This notebook walks through some simple examples of quantifying the apparent shift due to parallax of mountain terrain for the GOES-East (GOES-16) satellite for both individual points as well as larger areas.

This function finds the x and y coordinates in the ABI fixed rid coordinate system (azimuth (x) and elevation (y) scan angles in radians) given a point on the Earth's surface in Longitude and Latitude, with an elevation (z) relative to the reference ellipsoid.

```python
def LonLat2ABIangle(lon_deg, lat_deg, z, H, req, rpol, e, lon_0_deg):
    '''Find the ABI elevation (y) and scanning (x) angles (radians) of point P , given a latitude and longitude (degrees)'''
    
    # convert lat and lon from degrees to radians
    lon = np.radians(lon_deg)
    lat = np.radians(lat_deg)
    lon_0 = np.radians(lon_0_deg)
      
    # geocentric latitude
    lat_geo = np.arctan( (rpol**2 / req**2) * np.tan(lat) )

    # geocentric distance to point on the ellipsoid
    _rc = rpol / np.sqrt(1 - (e**2)*(np.cos(lat_geo)**2)) # this is rc if point is on the ellipsoid
    rc = _rc + z # this is rc if the point is offset from the ellipsoid by z (meters)

    # intermediate calculations
    Sx = H - rc * np.cos(lat_geo) * np.cos(lon - lon_0)
    Sy = -rc * np.cos(lat_geo) * np.sin(lon - lon_0)
    Sz = rc * np.sin(lat_geo)
    
    # calculate x and y scan angles
    y = np.arctan( Sz / Sx )
    x = np.arcsin( -Sy / np.sqrt( Sx**2 + Sy**2 + Sz**2 ) )
    
    # determine if this point is visible to the satellite
    condition = ( H * (H-Sx) ) < ( Sy**2 + (req**2 / rpol**2)*Sz**2 )
    if condition == True:
        #print('Point at {},{} not visible to satellite.'.format(lon_deg,lat_deg))
        return (np.nan, np.nan)
    else:
        return (x,y)
    
```

This function will compute the scan angle coordinates across a scene, given a DEM, taking parallax into account.

```python
def dem2ABIangle(dem,H,req,rpol,e,lon_0):

    abi_grid_x = np.zeros((dem.shape[1],dem.shape[2])) # make empty array to populate with values
    abi_grid_y = np.zeros((dem.shape[1],dem.shape[2])) # make empty array to populate with values

    # For each column:
    for ix in range(0,dem.shape[1]):
        # For each row:
        for iy in range(0,dem.shape[2]):

            # Print the column and row we're at
            print(ix,iy,end="\r")
            
            # Get the elevation from the DEM at this point
            z = dem[0,iy,ix]

            # If elevation value z is not "no data" value
            if z != -9999.0:
                
                lon = dem.x[ix] # get the longitude from the DEM at this point
                lat = dem.y[iy] # get the latitude from the DEM at this point
                
                # Calculate scan angles for a point above or below the ellipsoid surface
                abi_grid_x[iy,ix], abi_grid_y[iy,ix] = LonLat2ABIangle(lon,lat,z,H,req,rpol,e,lon_0)
        
        abi_grid = [abi_grid_x, abi_grid_y]
        
    return abi_grid
```
